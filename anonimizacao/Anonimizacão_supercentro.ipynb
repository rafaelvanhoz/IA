```python
# Se necessário, instale:
# !pip install pandas requests

import pandas as pd
import hashlib

# URL do arquivo
url = "C:\\Users\\rafae\\Downloads\\Pesquisa_Exame.csv"  # Substitua pelo caminho do seu arquivo local

# Nome do arquivo de saída
arquivo_saida = "pesquisa_exame_anonimizado.csv"

# Colunas que serão anonimizada
colunas_para_hash = ["CPF", "RGNumero", "NomPaciente"]


def aplicar_sha256(valor):
    """
    Converte o valor para string e aplica hash SHA-256.
    Mantém valores nulos como nulos.
    """
    if pd.isna(valor):
        return pd.NA
    
    valor_str = str(valor).strip()
    
    if valor_str == "":
        return pd.NA
    
    return hashlib.sha256(valor_str.encode("utf-8")).hexdigest()


try:
    # Tenta ler o CSV diretamente da URL
    df = pd.read_csv(url, sep=None, engine="python")
    
    # Verifica se as colunas existem
    colunas_existentes = [col for col in colunas_para_hash if col in df.columns]
    colunas_ausentes = [col for col in colunas_para_hash if col not in df.columns]
    
    if colunas_ausentes:
        print(f"Atenção: estas colunas não foram encontradas no arquivo: {colunas_ausentes}")
    
    # Aplica hash nas colunas encontradas
    for col in colunas_existentes:
        df[col] = df[col].apply(aplicar_sha256)
    
    # Salva o CSV anonimizado
    df.to_csv(arquivo_saida, index=False, encoding="utf-8-sig", sep=";")
    
    print(f"Arquivo salvo com sucesso: {arquivo_saida}")
    print("\nPrévia dos dados:")
    display(df.head())

except Exception as e:
    print("Erro ao ler o arquivo diretamente da URL.")
    print("Possível causa: o SharePoint pode exigir autenticação.")
    print(f"Detalhe do erro: {e}")
```
